## Q 15. Create a linear Regression Model which can predict CONVEX_AREA using the SHAPEFACTOR_1,	SHAPEFACTOR_2,	SHAPEFACTOR_3 and SHAPEFACTOR_4 columns. 

```python
df_new=df_fruit[['SHAPEFACTOR_1','SHAPEFACTOR_2', 'SHAPEFACTOR_3','SHAPEFACTOR_4','CONVEX_AREA']]
```

### Check Outliers

```python
col=['SHAPEFACTOR_1','SHAPEFACTOR_2', 'SHAPEFACTOR_3','SHAPEFACTOR_4','CONVEX_AREA']
```

```python
for i in col:
  plt.figure()
  sns.boxplot(y=i,data=df_new)
```

### Remove Outliers

```python
for c in col:
  per25=df_new[c].quantile(0.25)
  per75=df_new[c].quantile(0.75)
  iqr=per75-per25
  upperlimit= per75+1.5*iqr
  lowerlimit= per25-1.5*iqr
  df_new=df_new[(df_new[c]<upperlimit) & (df_new[c]>lowerlimit)]
  plt.figure()
  sns.boxplot(y=c,data=df_new)
```

```python
df_new
```

### Split dataset into Independant and Dependant Dataset

```python
x=df_new.iloc[:,:-1].values
y=df_new.iloc[:,-1].values
```

### Split the dataset into training data and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

### Scale the dataset

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
```

```python
x_test=sc.transform(x_test)
```

### Fit the Model

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

### Check Metrics

```python
from sklearn import metrics
```

```python
mse=metrics.mean_squared_error(y_test,y_pred)
```

```python
mse
```

```python
np.sqrt(mse)
```

```python
metrics.r2_score(y_test,y_pred)
```
